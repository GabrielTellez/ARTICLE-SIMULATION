```python
# Import packages to use
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import statsmodels.api as sm
import os
mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams.update({'errorbar.capsize': 2})
```

```python
# Define the file to explore
file='resumenLF1.csv'
N=100
s=1.0
# Import the data file
df=pd.read_csv(file, engine='python')
```

```python
df['nu-teo']=np.divide(-1.0,1.0+df['beta'])
dfp=df.copy()[(df['beta']==0.1) | (df['beta']==0.5) | (df['beta']==1.0) | (df['beta']==1.5) | (df['beta']==2.0) | (df['beta']==3.0) | (df['beta']==4.0) ]
dfs=df.copy()[(df['beta']==0.001) | (df['beta']==0.01) | (df['beta']==0.1) | (df['beta']==0.5) | (df['beta']==1.0) | (df['beta']==1.5) ]

```

```python
fig,ax=plt.subplots(figsize=(12,6))

ax.errorbar(dfp['beta'],dfp['nu'],yerr=dfp['nu_err'],
            fmt='o',label='Simulation data',color='blue')

betaplot=np.linspace(min(df['beta']),max(df['beta']),1000)
nuplot=np.divide(-1.0,1.0+betaplot)

ax.plot(betaplot,nuplot,label=r'$\nu\sim 1/(\beta+1)\qquad \beta \gg 1$',
        color='red',linestyle='--')

ax.set_xlabel(r'$\beta$',fontsize=30)
ax.set_ylabel(r'$\nu$',fontsize=30)
ax.set_title(r'$N=%i \quad \ell_f=%.1f$'%(N,s),fontsize=30)

plt.grid(alpha=0.4)
plt.legend(fontsize=24)
plt.show()
plt.close()
```

```python
fig,ax=plt.subplots(figsize=(12,6))

ax.errorbar(dfp['beta'],-dfp['nu'],yerr=dfp['nu_err'],
            fmt='o',label='Simulation data',color='blue')

betaplot=np.linspace(min(df['beta']),max(df['beta']),1000)
nuplot=np.divide(-1.0,1.0+betaplot)

ax.plot(betaplot,-nuplot,label=r'$\nu\sim 1/(\beta+1)\qquad \beta \gg 1$',
        color='red',linestyle='--')

ax.set_xlabel(r'$\beta$',fontsize=30)
ax.set_ylabel(r'$\nu$',fontsize=30)
ax.set_title(r'$N=%i \quad \ell_f=%.1f$'%(N,s),fontsize=30)

plt.grid(alpha=0.4)
plt.legend(fontsize=24,loc=2)
plt.show()
plt.close()
```

```python
fig,ax=plt.subplots(figsize=(12,6))
fig.patch.set_facecolor('white')

ax.errorbar(dfp['beta'],-dfp['nu'],yerr=dfp['nu_err'],
            fmt='o',label='Simulation data',color='blue')

betaplot=np.linspace(min(df['beta']),max(df['beta']),1000)
nuplot=np.divide(-1.0,1.0+betaplot)

ax.plot(betaplot,-nuplot,label=r'$\nu= 1/(\beta+1)\qquad \beta \gg 1$',
        color='red',linestyle='--')
ax.axhline(10,color='green',linestyle='-.',label=r'$\nu=0.5 \qquad \beta \ll 1$')
ax.annotate('(a)',fontsize=28,xy=(0.3,0.25))
ax.set_ylim(0.15,1.2)
ax.set_xlabel(r'$\beta$',fontsize=32)
ax.set_ylabel(r'$\nu$',fontsize=32)
ax.set_title(r'$N=%i \quad \ell_f=%.1f$'%(N,s),fontsize=30)
ax.grid(alpha=0.2)
ax.tick_params(axis='both', labelsize=12)
ax.legend(fontsize=21,loc=2,facecolor='white',framealpha=1)

#----------------------------------------------------------------
# Subplot to be embedded
ax2=fig.add_axes([.55, .52, .33, .33],facecolor='white')
ax2.errorbar(dfs['beta'],-dfs['nu'],yerr=dfs['nu_err'],
            fmt='o',color='blue')
ax2.axhline(0.5,color='green',linestyle='-.',label=r'$\nu=0.5$')
ax2.set_xscale('log')
ax2.annotate('(b)',fontsize=28,xy=(2e-3,0.35))
#ax2.set_xlabel(r'$\beta$',fontsize=22)
#ax2.set_ylabel(r'$\nu$',fontsize=22)
ax2.grid(alpha=0.2)
#----------------------------------------------------------------

plt.show()
plt.close()
```
